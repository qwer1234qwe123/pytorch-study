```python
list_data = [ [10, 20], [30, 40] ]
tensor1 = torch.Tensor(list_data)
print(tensor1)
print(f"tensor type: {type(tensor1)}, tensor shape: {tensor1.shape}")
print(f"tensor dtype: {tensor1.dtype}, tensor device: {tensor1.device}")
```

```python
if torch.cuda.is_available():
    tensor1 = tensor1.to("cuda")
print(f"tensor type: {type(tensor1)}, tensor shape: {tensor1.shape}")
print(f"tensor dtype: {tensor1.dtype}, tensor device: {tensor1.device}")
```

```python
import numpy as np

numpy_data = np.array(list_data)
tensor2 = torch.Tensor(numpy_data)
print(tensor2)
print(f"tensor type: {type(tensor2)}, tensor shape: {tensor2.shape}")
print(f"tensor dtype: {tensor2.dtype}, tensor device: {tensor2.device}")
```

```python
numpy_data = np.array(list_data)
tensor2_1 = torch.from_numpy(numpy_data)
print(tensor2_1)
print(f"tensor type: {type(tensor2_1)}, tensor shape: {tensor2_1.shape}")
print(f"tensor dtype: {tensor2_1.dtype}, tensor device: {tensor2_1.device}")
tensor2_2 = torch.from_numpy(numpy_data).float()
print('====================================')
print(tensor2_2)
print(f"tensor type: {type(tensor2_2)}, tensor shape: {tensor2_2.shape}")
print(f"tensor dtype: {tensor2_2.dtype}, tensor device: {tensor2_2.device}")
```

```python
# 텐서의 연산(Operations)
# 03/27
import math
a = torch.rand(1, 2) * 2 - 1
print(a)
print(torch.abs(a))
print(torch.ceil(a))
print(torch.floor(a))
print(torch.clamp(a, -0.5, 0.5))
```

```python
print(a)
print(torch.min(a))
print(torch.max(a))
print(torch.mean(a))
print(torch.std(a))
print(torch.prod(a))
print(torch.unique(torch.tensor([1, 2, 3, 1, 2, 2])))
```

```python
x = torch.rand(2, 2)
print(x)
y = torch.rand(2, 2)
print(y)
```

```python
print(x + y)
print(torch.add(x, y))
```

```python
result = torch.empty(2, 2)
torch.add(x, y, out=result)
print(result)
```

```python
print(x)
print(y)
y.add_(x)
print(y)
```

```python
print(x)
print(y)
print(x - y)
print(torch.sub(x, y))
print(x.sub(y))
```

```python
print(x)
print(y)
print(x * y)
print(torch.mul(x, y))
print(x.mul(y))
```

```python
print(x)
print(y)
print(x / y)
print(torch.div(x, y))
print(x.div(y))
```
